```python
from nltk.corpus import stopwords
english_stop_words = stopwords.words('english')
def remove_stop_words(corpus):
    removed_stop_words = []
    for review in corpus:
        removed_stop_words.append(
            ' '.join([word for word in review.split() 
                      if word not in english_stop_words])
        )
    return removed_stop_words
```

```python
#We will limit the maximum review length to max_words by truncating longer reviews and padding shorter reviews with a null value (0). We can accomplish this using the pad_sequences() function in Keras. For now, set max_words to 500.
from keras.preprocessing import sequence
max_words = 200
word_padding = sequence.pad_sequences(word_id, maxlen=max_words)
#x_test = sequence.pad_sequences(x_test, maxlen=max_words)
```

```python
word_padding
```

```python
vocabulary_size = (word_padding.max()+1).astype('int64')
from keras import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout
embedding_size=32
model=Sequential()
model.add(Embedding(vocabulary_size, 32))
model.add(LSTM(64,dropout=0.4,recurrent_dropout=0.4,return_sequences=True))
model.add(LSTM(32,dropout=0.5,recurrent_dropout=0.5,return_sequences=False))
model.add(Dense(5, activation='softmax'))
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
batch_size = 64
num_epochs = 10
x_valid, y_valid = word_padding[:batch_size], Priority_cat[:batch_size]
x_train2, y_train2 = word_padding[batch_size:], Priority_cat[batch_size:]
model.fit(x_train2, y_train2,validation_data=(x_valid,y_valid) ,batch_size=batch_size, epochs=num_epochs)
```
